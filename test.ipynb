```python
from DPInvisibleDecay import DPInvisibleDecay
darkSHINE = DPInvisibleDecay(3e14, 8, verbose=1)
darkSHINE.initialization()
X,Y,Z = darkSHINE.GetGrid()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import interp2d

interp_func = interp2d(X, Y, Z, kind='cubic')
x_new = np.linspace(-3, 0, 1000)
y_new = np.linspace(-7, -2, 1000)
z_new = interp_func(x_new, y_new)

plt.rc('text', usetex=True)
# plt.contour(X, Y, Z, levels=[0.9])
plt.contour(x_new, y_new, z_new, levels=[0.9])
plt.xlabel('$log_{10}(m_A)$')
plt.ylabel('$log_{10}\epsilon$')
plt.show()
```
